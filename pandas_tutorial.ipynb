# 색인 (indexing)

```python
train["Sex"]=="male"
```

```python
train[train["Sex"]=="male"].head(3)
```

성별이 male인 데이터를 가져온다.
train["Sex"]=="male"은 성별이 male이면 True, 아니면 False
이 리스트를 train[]안에 넣으면 값이 True인 데이터만 가져오는데 
이를 판다스 indexing이라 한다.

```python
train[train["Age"]>=60].head(10)
```

```python
train[train["Embarked"]=="S"].head(3)
```

```python
train[train["Sex"]!="male"].head(4)
```

isin()함수를 사용. 이 함수에는 리스트 값이 들어가는데 
리스트 안에 있는 값들 중 하나라도 있으면 True, 아니면 False 반환

```python
train[train["Embarked"].isin(["C","S"])].head(3)
```

str.contains
문자열 컬럼에서 자주 사용, 문자열 컬럼에 특정 문자열이 포함되어 있을 경우에는 True,
아닐 경우에는 False

```python
train[train["Ticket"].str.contains("STON")].head(3)
```

```python
train[train["Name"].str.contains("o")].head()
```

isnull() 함수
컬럼 안의 내용이 비어있는지의 여부 확인(비어있는 데이터 = NaN)
NaN : Not a Number 혹은 null

```python
train[train["Age"].isnull()].head(3)
```

notnull() 함수
비어있지 않으면 True, 비어있으면 Flase


```python
train[train["Age"].notnull()].head(3)
```

~

```python
train[~(train["Pclass"] == 1)].tail(3)
```

>indexing을 할 때에 여러 조건을 동시에 넣을 수 있다.
& (and)
| (or)
단, 파이썬 문법인 and, or는 판다스에서 동작하지 않음.

```python
train[(train["Sex"]=="female") & (train["Pclass"]==1)].head(4)
```

```python
train[(train["Age"].isnull()) & (train["Embarked"]=="Q")].head(2)
```

```python
# 위와 같음
A = train["Age"].isnull()
B = train["Embarked"]=="Q"
train[A&B].head(2)
```

# Crosstab

판다스에서 제공하는 간단한 통계분석 기능

>첫 번째 인자(index)와 두 번째 인자(Columns)에 각각 컬럼명을 넣어주면, 두 컬럼에 해당하는 데이터의 발생 빈도를 출력

```python
pd.crosstab(index=train["Embarked"], columns=train["Survived"])
# C, Q, S마다 생존자(1)와 사망자(0)의 인원수가 출력
```

>index나 columns에는 여러 개의 컬럼을 넣을 수 있다.

```python
pd.crosstab(index=[train["Pclass"], train["Sex"]], columns=train["Survived"])
```
